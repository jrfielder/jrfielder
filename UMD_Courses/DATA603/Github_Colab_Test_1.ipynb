```python
import numpy as np
import matplotlib.pyplot as plt
x = np.array([0,.5,1.,1.5,2.,2.5,3.])
y = np.sin(x)
plt.figure()
plt.plot(x,y, label='sin(x)')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.title('not very good plot of sine')
plt.grid()
plt.legend()
```

```python
no_samples = 1000
x = np.linspace(0,2*np.pi, no_samples)
y_sin = np.sin(x)
y_cos = np.cos(x)
plt.figure()
plt.plot(x,y_sin, label='sin(x)')
plt.plot(x,y_cos, label='cos(x)')
plt.xlabel('x')
plt.ylabel('sin/cos(x)')
plt.title(' a better plot of sine and cosine')
plt.grid()
plt.legend()
```
